Get a list of top 100 artists from Billboard.com for the period 2010 - 2018 and create a histogram to visualize number of repeated appreances for top artists year-over-year.

We then query the Spotify API to get the top 10 tracks for the list of top artists and perform an EDA on the audio features for tracks.

```python
#scrape the top 100 artists from Billboard.com for a particular year.

#install dependencies
!pip install bs4
from bs4 import BeautifulSoup
import requests

artist_list = []
artist_list_clean = []
def topartists(page_url):
    url = page_url
    html = requests.get(url)
    #get html content for the page requested
    html_content = BeautifulSoup(html.content, 'html.parser')
    #find the artist div class
    artist_div = html_content.findAll('div', class_="ye-chart-item__title")
    for item in artist_div:
        #get the artist name from their link text 
        artist_list.append(item.a.text)
    for item in artist_list:
        artist_list_clean.append(item.strip())
    return artist_list_clean
```
